```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

url = 'https://gist.githubusercontent.com/zgulde/018f5d601efc79cb741a7aba92f51d29/raw/c08f1c09f605b43e9a354890f9dd68f296e76dc4/students.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
plt.figure(figsize=(13, 7))
plt.rcParams.update({'font.size': 16})
plt.scatter(x=df.syntax_errors, y=df.ml_coffee)
plt.title('Relationship of Syntax Errors to Coffee (ml)')
plt.xlabel('Syntax Errors - log scale')
plt.ylabel('Coffee (ml)')
plt.show()
# not sure if need to convert or highlight x axis in log 10?
# maybe convert ml to liters?
```

```python
# WALKTHROUGH
#reverse axis
plt.figure(figsize=(13, 7))
plt.rcParams.update({'font.size': 16})
plt.scatter(x=df.ml_coffee, y=df.syntax_errors)
plt.title('Students who drink more coffee make more syntax error)')
plt.xlabel('Syntax Errors in Billions')
plt.ylabel('Coffee (ml)')
plt.show()
```

```python
df['syntax_errors_billions'] = df.syntax_errors/1000000000
df['coffee_cups'] = df.ml_coffee/30/8
```

```python
plt.figure(figsize=(13, 7))
plt.rcParams.update({'font.size': 16})
plt.scatter(x=df.coffee_cups, y=df.syntax_errors_billions)
plt.title('Students who drink more coffee make more syntax error)')
plt.xlabel('Syntax Errors in Billions')
plt.ylabel('Coffee in cups')
plt.show()
```
